```python
import pandas as pd
```

```python
from selenium import webdriver
driver = webdriver.Chrome()
```

```python
driver.get('https://www.police.gov.hk/ppp_en/03_police_message/pr/pr_archives.html?month=201910')
```

```python
from selenium.webdriver.support.ui import Select

# select = Select(driver.find_elements_by_name('h3'))

items = driver.find_elements_by_class_name('acc')

rows = []

for item in items:
  print(item.text)

  rows.append(item.text)

  # row['date'] = item.find(class_='acc').text.strip()
  # row['detail'] = item.find(class_='acc-content').text.strip()

  print(rows)
len(rows)
```

```python
items = driver.find_elements_by_tag_name('h3')

rows = []

for item in items:
  # print(item.text)
  
  row = {}


  row['case'] = item.text.strip()


  rows.append(row)

  # row['date'] = item.find(class_='acc').text.strip()
  # row['detail'] = item.find(class_='acc-content').text.strip()

len(rows)
print(rows)
```

```python
df = pd.DataFrame(rows)
df.head(15)
```

```python
df = df[12:]
df.head()
```

```python
df['date'] = df.case.str.extract("(.*\d).*", expand=False)
df['case'] = df.case.str.extract(".*\d(.*)", expand=False)
```

```python
df = df[['date', 'case']]
```

```python
df.head()
```

```python
df.to_csv("scraped.csv", index=False)
```
